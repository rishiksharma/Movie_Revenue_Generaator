```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import requests
from bs4 import BeautifulSoup
from collections import OrderedDict
import pandas
import time
```

```python
'''
Get_MainList - func() 

Scrapes the [start_url] using the Selenium WebDriver [chrome_driver] for the list of movie names and associated 
data and parameters like 'Release Data',etc. as well as Urls to each movie's details.
The Urls will be used in subsequent sections to scrape for Metrics and Cast-Crew related information.

'''
def Get_MainList(chrome_driver,start_url):

    chrome_driver.get(start_url)
    df_mn = pandas.DataFrame()
    print(start_url)
    main_div = chrome_driver.find_element_by_id('main')
    tabular_data = main_div.find_element_by_tag_name('table')
    table_rows = tabular_data.find_elements_by_tag_name('tr')

    for each_row in table_rows:
        text_data = each_row.find_elements_by_class_name('data')
        link_data = each_row.find_elements_by_tag_name('a')
        lst_row = []
        for t_d in text_data:
            lst_row.append(t_d.text)
        for l_d in link_data:
            lst_row.append(l_d.text)
            lst_row.append(l_d.get_attribute('href'))
        #print(lst_row)
    
        if(len(lst_row) > 0):
            pd_row = pandas.Series(lst_row)
            #print(pd_row)
            df_mn = df_mn.append(pd_row,ignore_index=True)

    return df_mn
```

```python
'''
Place from where the actual work begins....
'''

origin_url = 'https://www.the-numbers.com/movie/budgets/all'
df_main = pandas.DataFrame()

chrome_driver = webdriver.Chrome(ChromeDriverManager().install())

for pg in range(1,501,100):
    
    pg_str = origin_url
    if(pg > 100)
        pg_str = '{1}/{0}'.format(pg,origin_url)

    # usage of the Get_MainList method to generate the list of movies
    df_m = Get_MainList(chrome_driver, pg_str)
    df_main = pandas.concat([df_main.reset_index(drop=True),df_m.reset_index(drop=True)],axis=0)

chrome_driver.quit()
```

```python
def GetCast_Numbers(chrome_driver,tab_url):
    
    # url format should be as: 'https://www.the-numbers.com/movie/Avengers-Endgame-(2019)?#tab=cast-and-crew'
    chrome_driver.get(tab_url) 
    # Movie Name Eg. Avengers-Endgame-(2019)
    time.sleep(5)
    cAstaNdcRew = chrome_driver.find_element_by_id('cast-and-crew')
    # cAstAndcRew is the massive 'div' that holds all other casting 'div's.
    # The inner 'div's are for 1. Main Cast 2. Supporting Cast 3. Cameos 4. Production and Technical assistance amongst others          we are not very interested in.
    # Within the specific cast 'div's there are tables prefaced with headings. These tables are a list of actors
    # Each actor name is in a 'span' in an 'anchor' tag in a 'td'

    dctNames = {}
    try:
        actorTypes = cAstaNdcRew.find_elements_by_class_name('cast_new')
        grpNo = 1
        for actorType in actorTypes:# Each of Main Cast, Cameo, Supporting etc.

        # Initialize a list to hold the actors in this section
            actorNames = []
        # Section Name / Actor Type
        #actorType_Name = (actorType.find_elements_by_tag_name('h1'))[0].text
        #actorType_Name = actorType_Name.replace(' ','_')
            actorType_Name = "Artist_Group_" + str(grpNo)
        # Get all values from anchor tags in the section
            actor_Names = actorType.find_elements_by_tag_name('a')
       
            if len(actor_Names) > 0:
                for actor_Name in actor_Names:
                 # print('a: ',actor_Name.text) # Crude Debugging
                    actorNames.append(actor_Name.text)
        
            dctNames[actorType_Name] = actorNames
            grpNo += 1
    except Exception:
        print("Exception caught")
    finally:
        df = pandas.DataFrame.from_dict(dctNames,orient='index')  
        return df
```

```python
def GetCastMetrics(dfc):
    
    dfnew = dfc.head(1)
    for i in range(len(dfc.columns)):
        dfnew[dfc.columns[i] + '_Count'] = dfc[dfc.columns[i]].count()

    return dfnew
```
